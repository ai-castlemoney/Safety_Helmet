```python
import pandas as pd

case = pd.read_csv('final_case_EDA_utf8.csv', header = 0, index_col = 0)
case.head()
```

```python
# 세부적으로 컬럼 분할
case['공종1'] = case['공종'].str.split(' > ').str[0]
case['공종2'] = case['공종'].str.split(' > ').str[1]

case['객체분류'] = case['사고객체'].str.split(' > ').str[0]
case['객체종류'] = case['사고객체'].str.split(' > ').str[1]

case['시설장소'] = case['장소'].str.split(' / ').str[0]
case['위치장소'] = case['장소'].str.split(' / ').str[1]

case['부위종류'] = case['부위'].str.split(' / ').str[0]
case['세부위치'] = case['부위'].str.split(' / ').str[1]

# 분할한 원 컬럼은 제거
case = case.drop(['공종', '사고객체', '장소', '부위'], axis = 1)

case.head()
```

```python
import numpy as np

case = case.replace(np.nan, '')
```
