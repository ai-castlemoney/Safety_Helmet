```python
import requests
from bs4 import BeautifulSoup
from selenium import webdriver

# 크롬 브라우저 열기
browser = webdriver.Chrome('./chromedriver.exe')
browser.maximize_window()  # 창 최대화

# 2. 스크래핑 진행
def create_soup(url):
    res = requests.get(url)
    res.raise_for_status()
    soup = BeautifulSoup(res.text, 'lxml')
    return soup

# 1. 코샤 국내재해사례 페이지로 이동
title_list = []
sentence_list = []

for i in range(3):
    url = 'https://www.kosha.or.kr/kosha/data/construction.do?mode=list&&articleLimit=10&article.offset={}0'.format(i)
    browser.get(url)

    # 제목 텍스트 추출
    soup = create_soup(url)
    lists = soup.find_all('td', attrs={'class':'board-list-title'})

    for idx, list in enumerate(lists):
        title = list.find('a').get_text().strip()
        title_list.append(title)
        print('{}. {}'.format(idx+1, title))

    for j in range(i*10, len(title_list)):
        browser.find_element_by_link_text(title_list[j]).click()
        elem = browser.find_element_by_class_name('pre')
        sentence_list.append(elem.text)
        browser.back()
        
print(title_list)
print(sentence_list)
```
