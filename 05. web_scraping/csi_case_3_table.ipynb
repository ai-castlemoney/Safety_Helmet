```python
import requests
from bs4 import BeautifulSoup
import re

url = 'https://www.csi.go.kr/acd/acdCaseView.do?case_no=13048'
res = requests.get(url)
res.raise_for_status()   # 문제 있으면 프로그램 종료

soup = BeautifulSoup(res.text, 'lxml')
lists = soup.find_all('div', attrs = {'class' : 'contentsRow'})

key_list = ['사고명', '발생일시', '사고인지 시간', '공공/민간 구분', '기상상태', '시설물 종류', '인적사고', '보호(방호)조치여부', '물적사고', '공종', '사고객체', '작업프로세스', '장소', '부위', '사고경위', '사고원인', '구체적 사고원인', '사망자수(명)', '부상자수(명)', '피해금액', '피해내용', '사고신고사유', '사고발생후 조치사항', '재발방지대책', '공사종류', '공사비', '낙찰률', '공사기간', '공정률', '작업자수', '안전관리계획', '설계안전성검토', '사고조사방법', '위원회조사필요성', '위원회구성(안)', '향후조치계획']
value_list = []

for list in lists:
    contents = list.find_all('td', attrs = {'class':'t-left'})
    for content in contents:
        content = content.text.replace('\n', '')
        content = content.replace('\t', '')
        content = content.replace('\r', '')
        value_list.append(content)

dictionary = dict(zip(key_list, value_list))
print(dictionary)

# print(key_list)
# print(value_list)

print('key 갯수 : ', len(key_list))
print('value 갯수 : ', len(value_list))
```
