## **1. 데이터 불러오기**
---

```python
import pandas as pd

case = pd.read_csv('EDA_accidentcase.csv', header = 0, index_col = 0)
case.head()
```

## **2. 컬럼 분할**
---

```python
# 세부적으로 컬럼 분할
case['공종대분류'] = case['공종'].str.split(' > ').str[0]
case['공종소분류'] = case['공종'].str.split(' > ').str[1]

case['객체대분류'] = case['사고객체'].str.split(' > ').str[0]
case['객체소분류'] = case['사고객체'].str.split(' > ').str[1]

case['시설장소'] = case['장소'].str.split(' / ').str[0]
case['위치장소'] = case['장소'].str.split(' / ').str[1]

case['부위내용'] = case['부위'].str.split(' / ').str[0]
case['부위위치'] = case['부위'].str.split(' / ').str[1]

# 분할한 원 컬럼은 제거
case = case.drop(['공종', '사고객체', '장소', '부위'], axis = 1)

case.head()
```

## **3. '객체소분류', '부위내용' 컬럼 비교**
---

```python
# 각 컬럼 값 확인
case[['객체소분류', '부위내용']].head(30)
```

```python
# 컬럼 내 값이 동일한지 확인
print(case['객체소분류'][1] == case['부위내용'][1])
print(case['객체소분류'][1], case['부위내용'][1])
```

```python
# 컬럼의 결측치 확인
case[['객체소분류', '부위내용']].isnull().sum()
```

```python
case[['객체소분류']].nunique()
```

```python
# 각 컬럼 내 값들이 서로 모두 동일한지 확인

# 컬럼을 리스트로 반환
case_one = case['객체소분류'].tolist()
case_two = case['부위내용'].tolist()

# 2개의 컬럼의 값이 다를 경우, 출력
for i in range(len(case_one)):
    if case_one[i] != case_two[i]:
        print(i, case_one[i], case_two[i])
```

```python
import numpy as np

# nan 값을 ''로 대체 
case = case.replace(np.nan, '')
```

```python
# 다시 재확인
case_one = case['객체소분류'].tolist()
case_two = case['부위내용'].tolist()

for i in range(len(case_one)):
    if case_one[i] != case_two[i]:
        print(i, case_one[i], case_two[i])
```

### **[결론]**
- '객체소분류'와 '부위내용'의 값이 모두 동일함을 확인함
- 둘 중 하나 제거 필요 ('부위내용' 컬럼 제거)

```python
case = case.drop(['부위내용'], axis = 1)
case
```
