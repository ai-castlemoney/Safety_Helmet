## **1. 기본 사용**

```python
import matplotlib.pyplot as plt
import squarify

plt.style.use('default')
plt.rcParams['figure.figsize'] = (4, 4)
plt.rcParams['font.size'] = 12

sizes = [40, 30, 5, 25]

squarify.plot(sizes)
plt.show()
```

```python
plt.style.use('default')
plt.rcParams['figure.figsize'] = (4, 4)
plt.rcParams['font.size'] = 12

sizes = [40, 30, 5, 25]

squarify.plot(sizes, 10, 10)  # dx, dy를 함께 입력 : 사각형 넓이의 총합이 dx*dy됨
plt.show()
```

## **2. 레이블 표시**

```python
plt.style.use('default')
plt.rcParams['figure.figsize'] = (4, 4)
plt.rcParams['font.size'] = 12

sizes = [40, 30, 5, 25]
labels = ['A', 'B', 'C', 'D']   # 레이블 표시

squarify.plot(sizes, 10, 10, label = labels)
plt.show()
```

## **3. 색상 지정하기**
- 색상 종류 : https://codetorial.net/matplotlib/set_color.html

```python
plt.style.use('default')
plt.rcParams['figure.figsize'] = (4, 4)
plt.rcParams['font.size'] = 12

sizes = [40, 30, 5, 25]
labels = ['A', 'B', 'C', 'D']
colors = ['lightgreen', 'cornflowerblue', 'mediumpurple', 'lightcoral']   # 색상 선정

squarify.plot(sizes, 10, 10, label = labels, color = colors)
plt.show()
```

## **4. 축 제거하기**

```python
plt.style.use('default')
plt.rcParams['figure.figsize'] = (4, 4)
plt.rcParams['font.size'] = 12

sizes = [40, 30, 5, 25]
labels = ['A', 'B', 'C', 'D']
colors = ['lightgreen', 'cornflowerblue', 'mediumpurple', 'lightcoral']

squarify.plot(sizes, 10, 10, label = labels, color = colors)
plt.axis('off')  # 축 제거
plt.show()
```

## **5. 테두리 두께 정하기**

```python
sizes = [40, 30, 5, 25]
labels = ['A', 'B', 'C', 'D']
colors = ['lightgreen', 'cornflowerblue', 'mediumpurple', 'lightcoral']

squarify.plot(sizes, 10, 10, label = labels, color = colors, bar_kwargs=dict(linewidth=3, edgecolor='#eee'))
plt.axis('off')
plt.show()
```

## **6. Numpy 활용**

```python
import numpy as np

np.random.seed(0)

sizes = np.random.randint(1, 100, size = 50)

squarify.plot(sizes, bar_kwargs=dict(linewidth=1, edgecolor='#eee'))
plt.axis('off')
plt.show()
```
