```python
import matplotlib.pyplot as plt
import seaborn as sns
from matplotlib import font_manager, rc
import pandas as pd
import numpy as np

font_name = font_manager.FontProperties(fname="c:/Windows/Fonts/malgun.ttf").get_name()
rc('font', family=font_name)
```

```python
case = pd.read_csv('2. EDA_원인분석, 고유값.csv', header = 0, index_col = 0)
```

# **1. 인적사고**
---

```python
case['인적사고'].unique()
```

```python
case['인적사고'] = case['인적사고'].str.split('(').str[0]
```

```python
case['인적사고'].unique()
```

```python
# nan 값을 ''로 대체 
case['인적사고'] = case['인적사고'].replace(np.nan, '없음')
```

```python
case.to_csv('인적, 물적사고.csv')
```

```python
colors = sns.color_palette('Dark2')
fig = plt.figure(figsize=(10,5))
fig.set_facecolor('white')

ax = fig.add_subplot()

sns.countplot(x = '인적사고', data = case, order = case['인적사고'].value_counts().index, palette=colors)
plt.xticks(rotation = 45)
plt.xlabel('')

plt.title('인적사고', fontsize = 20)
plt.show()
```

# **2. 물적사고**
---

```python
case['물적사고'].unique()
```

```python
# nan 값을 ''로 대체 
case['물적사고'] = case['물적사고'].replace(np.nan, '없음')
```

```python
fig = plt.figure(figsize=(10,5))
fig.set_facecolor('white')

ax = fig.add_subplot()

sns.countplot(x = '물적사고', data = case, order = case['물적사고'].value_counts().index)
plt.xticks(rotation = 45)
plt.xlabel('')

plt.title('물적사고', fontsize = 20)
plt.show()
```

```python
colors = sns.color_palette('Dark2')
fig = plt.figure(figsize=(10,5))
fig.set_facecolor('white')

ax = fig.add_subplot()

sns.countplot(x = '물적사고', data = case, order = case['물적사고'].value_counts().index, palette=colors)
plt.xticks(rotation = 45)
plt.xlabel('')

plt.title('물적사고', fontsize = 20)
plt.show()
```

```python
# '없음' 값 제거
# 참고 : https://bohemihan.tistory.com/entry/Pandas-%ED%8A%B9%EC%A0%95%EC%A1%B0%EA%B1%B4-%EB%A7%8C%EC%A1%B1%ED%95%98%EB%8A%94-%EC%9D%B8%EB%8D%B1%EC%8A%A4%ED%96%89-%EC%B9%BC%EB%9F%BC%EC%97%B4-%EC%82%AD%EC%A0%9C%ED%95%98%EA%B8%B0

re_case = case[case['물적사고'].str.contains('없음')].index
case.drop(re_case, inplace=True)
```

```python
fig = plt.figure(figsize=(10,5))
fig.set_facecolor('white')

ax = fig.add_subplot()

sns.countplot(x = '물적사고', data = case, order = case['물적사고'].value_counts().index)
plt.xticks(rotation = 45)
plt.xlabel('')

plt.title('물적사고(없음 제거)', fontsize = 20)
plt.show()
```

```python
colors = sns.color_palette('Set1')
fig = plt.figure(figsize=(10,5))
fig.set_facecolor('white')

ax = fig.add_subplot()

sns.countplot(x = '물적사고', data = case, order = case['물적사고'].value_counts().index, palette=colors)
plt.xticks(rotation = 45)
plt.xlabel('')

plt.title('물적사고', fontsize = 20)
plt.show()
```
