```python
import requests
from bs4 import BeautifulSoup
from selenium import webdriver
import time

# 크롬 브라우저 열기
browser = webdriver.Chrome('./chromedriver.exe')
browser.maximize_window()  # 창 최대화

# 1. 재해사례 사이트로 이동
url = 'https://www.csi.go.kr/acd/acdCaseList.do'
browser.get(url)

# 2. 목록 타이틀 스크래핑 진행
def create_soup(url):
    res = requests.get(url)
    res.raise_for_status()
    soup = BeautifulSoup(res.text, 'lxml')
    return soup

soup = create_soup(url)
lists = soup.find_all('tr', attrs = {'class' : 'odd gradeX'})

for idx, list in enumerate(lists):
    title = list.find('td', attrs = {'class':'t-left'})
    print(idx+1, title.get_text().strip())

for i in range(2, 11):
    # 3. 다음 목록 페이지로 넘어가기
    java = "goPage('{}');".format(i)
    browser.execute_script(java)

    # 4. 타이틀 스크래핑 진행
    html = browser.page_source
    soup = BeautifulSoup(html,'html.parser')
    lists = soup.find_all('tr', attrs = {'class' : 'odd gradeX'})

    for idx, list in enumerate(lists):
        title = list.find('td', attrs = {'class':'t-left'}).get_text().strip()
        print('{}. {}'.format(idx+1+(i-1)*10, title))
```
