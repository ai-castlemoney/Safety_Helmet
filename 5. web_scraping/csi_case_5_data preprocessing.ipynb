```python
import pandas as pd

case = pd.read_csv('final_case_re_utf8.csv', header = 0, index_col = 0)
case.head()
```

```python
# 내국인/외국인 열로 나누기
case['사망자_내국인'] = case['사망자수(명)'].str.split(':').str[1]
case['사망자_외국인'] = case['사망자수(명)'].str.split(':').str[2]

case['부상자_내국인'] = case['부상자수(명)'].str.split(':').str[1]
case['부상자_외국인'] = case['부상자수(명)'].str.split(':').str[2]

# 사망자수, 부상자수 열 제거
case = case.drop(['사망자수(명)', '부상자수(명)'], axis = 1)

case.head()
```

```python
case['사망자_내국인'] = case['사망자_내국인'].str.extract('(\d+)')
case['사망자_외국인'] = case['사망자_외국인'].str.extract('(\d+)')

case['부상자_내국인'] = case['부상자_내국인'].str.extract('(\d+)')
case['부상자_외국인'] = case['부상자_외국인'].str.extract('(\d+)')

case.head()
```

```python
# 문자형인 컬럼을 숫자형으로 변경
case[['사망자_내국인', '사망자_외국인', '부상자_내국인', '부상자_외국인']] = case[['사망자_내국인', '사망자_외국인', '부상자_내국인', '부상자_외국인']].apply(pd.to_numeric)
```

```python
# 내국인 수와 외국인 수를 더함
case['사망자수'] = case['사망자_내국인'] + case['사망자_외국인']
case['부상자수'] = case['부상자_내국인'] + case['부상자_외국인']

# 필요없는 컬럼 제거
case = case.drop(['사망자_내국인', '사망자_외국인', '부상자_내국인', '부상자_외국인'], axis = 1)

case.head()
```

```python
df = pd.DataFrame(case)

df.to_csv('final_case_r1.csv')
```
